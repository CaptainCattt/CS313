#THÊM và ĐỌC dữ liệu của model

```python
import pickle
# Load mô hình
loaded_model = pickle.load(open('random_forest_model.pkl', 'rb'))
```

#Input dữ liệu để dự đoán

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Hàm chuyển đổi điểm thành khoảng phân loại
def chuyen_doi_diem(diem):
    if diem >= 8:
        return 'A'
    elif diem >= 6.5:
        return 'B'
    elif diem >= 5:
        return 'C'
    else:
        return 'F'

# Hàm để yêu cầu người dùng nhập dữ liệu
input_data = {}
input_data['IT001'] = float(input('Nhập giá trị cho IT001: '))
input_data['IT003'] = float(input('Nhập giá trị cho IT003: '))
input_data['IT004'] = float(input('Nhập giá trị cho IT004: '))
input_data['IT005'] = float(input('Nhập giá trị cho IT005: '))
input_data['IT007'] = float(input('Nhập giá trị cho IT007: '))
input_data['MA003'] = float(input('Nhập giá trị cho MA003: '))
input_data['MA004'] = float(input('Nhập giá trị cho MA004: '))
input_data['MA005'] = float(input('Nhập giá trị cho MA005: '))
input_data['MA006'] = float(input('Nhập giá trị cho MA006: '))

# Tạo DataFrame từ dữ liệu nhập vào
data = pd.DataFrame(input_data, index=[0])

# Áp dụng hàm chuyển đổi cho từng phần tử trong DataFrame
data = data.applymap(chuyen_doi_diem)
label_encoder = LabelEncoder()
label_encoder.fit(['A', 'B', 'C','F'])  # Fit LabelEncoder với các giá trị có thể của các cột
data['IT001'] = label_encoder.transform(data['IT001'])
data['IT003'] = label_encoder.transform(data['IT003'])
data['IT004'] = label_encoder.transform(data['IT004'])
data['IT005'] = label_encoder.transform(data['IT005'])
data['IT007'] = label_encoder.transform(data['IT007'])
data['MA003'] = label_encoder.transform(data['MA003'])
data['MA004'] = label_encoder.transform(data['MA004'])
data['MA005'] = label_encoder.transform(data['MA005'])
data['MA006'] = label_encoder.transform(data['MA006'])

result = loaded_model.predict(data)
result = pd.Series(result).replace({0: 'Với điểm số trên bạn có thể sẽ trượt môn', 1: 'Với điểm số trên bạn có khả năng sẽ qua môn'})

# In kết quả dự đoán
print('\n\n\nKết quả dự đoán:', result[0])
```
