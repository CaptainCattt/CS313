```python
# Tính điểm trung bình học kì của một sinh viên
def diem_tb_hk_nh(masv, hocky, namhoc, diem):
    # Lọc các môn học của sinh viên trong học kỳ và năm học tương ứng
    filtered_data = diem[(diem['masv'] == masv) & (diem['hocky'] == hocky) & (diem['namhoc'] == namhoc)]
    
    # Lọc các môn học có trạng thái học tập bình thường, trả nợ hoặc cải thiện
    filtered_data = filtered_data[filtered_data['trangthai'].isin([1, 2, 3])]
    
    # Loại bỏ các môn học có số tín chỉ bằng 0
    filtered_data = filtered_data[filtered_data['sotc'] > 0]
    
    # Tính tổng số tín chỉ và tổng điểm của các môn học
    tong_sotc = filtered_data['sotc'].sum()
    tong_diem = 0
    
    for index, row in filtered_data.iterrows():
        tong_diem += float(row['diem']) * row['sotc']
    
    diem_tb_hk_nh = tong_diem / tong_sotc
    return diem_tb_hk_nh
```

```python
df_02_diem['diem_tb_hk'] = df_02_diem.groupby(['masv', 'namhoc', 'hocky']).apply(lambda x: diem_tb_hk_nh(x.name[0], x.name[2], x.name[1], x)).reset_index(level=[0, 1, 2], drop=True)
df_02_diem
```

```python
# Tính điểm trung bình toàn khóa của một sinh viên 
def diem_tb_tk(masv, hocky, namhoc, diem):
    # Lọc các môn học của sinh viên trong học kỳ và năm học tương ứng
    filtered_data = diem[(diem['masv'] == masv) & (diem['hocky'] == hocky) & (diem['namhoc'] == namhoc)]
    
    # Lọc các môn học có trạng thái học tập bình thường, trả nợ hoặc cải thiện
    filtered_data = filtered_data[filtered_data['trangthai'].isin([1])]
    
    # Loại bỏ các môn học có số tín chỉ bằng 0
    filtered_data = filtered_data[filtered_data['sotc'] > 0]
    
    # Tính tổng số tín chỉ và tổng điểm của các môn học
    tong_sotc = filtered_data['sotc'].sum()
    tong_diem = 0
    
    for index, row in filtered_data.iterrows():
        tong_diem += float(row['diem']) * row['sotc']
    
    diem_tb_tk = tong_diem / tong_sotc
    return diem_tb_tk
```

```python
df_02_diem['diem_tb_tk'] = df_02_diem.groupby(['masv', 'namhoc', 'hocky']).apply(lambda x: diem_tb_tk(x.name[0], x.name[2], x.name[1], x)).reset_index(level=[0, 1, 2], drop=True)
df_02_diem
```
