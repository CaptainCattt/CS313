##1. LÀM SẠCH DỮ LIỆU

```python
import pandas as pd
import numpy as np
```

```python
df_14_totnghiep = pd.read_csv("/content/drive/MyDrive/Khai thác dữ liệu và ứng dụng - CS313.N21/Education dataset/14.totnghiep.csv",
                              encoding='unicode-escape')
```

```python
df_14_totnghiep.head(10)
```

```python
df_14_totnghiep.info()
```

```python
# Xóa cột id
df_14_totnghiep = df_14_totnghiep.iloc[:, 1:]

# Chuẩn hóa tên cột (bỏ khoảng trắng trước tên cột)
df_14_totnghiep = df_14_totnghiep.rename(columns = {' masv': 'masv',
                                                    ' xeploai': 'xeploai',
                                                    ' soquyetdinh': 'soquyetdinh',
                                                    ' ngaycapvb': 'ngaycapvb'})
# Kiểm tra
df_14_totnghiep.columns
```

```python
# Xóa 2 dòng không chứa dữ liệu
df_14_totnghiep.dropna(axis=0, inplace=True)
```

```python
# Chia cột soquyetdinh thành 2 cột phân cách bởi "/"
df_14_totnghiep[['soquyetdinh','chitiet']] = df_14_totnghiep['soquyetdinh'].str.split("/", 1, expand=True)
                                                                                      
# Chỉ lấy phần số trước "/", bỏ phần sau
df_14_totnghiep.drop(columns=['chitiet'], inplace=True)

# Kiểm tra
df_14_totnghiep['soquyetdinh'].unique()
```

```python
# Chuẩn hóa các trường hợp ngoại lệ còn sót lại
df_14_totnghiep.replace(' 443-QD-DHCNTT', '443', inplace=True)
df_14_totnghiep.replace(' 519-QD-DHCNTT', '519', inplace=True)

# Chuyển cột soquyetdinh sang kiểu dữ liệu int
df_14_totnghiep['soquyetdinh'] = df_14_totnghiep['soquyetdinh'].astype(int)

# Kiểm tra
df_14_totnghiep['soquyetdinh'].unique()
```

```python
# Cột xeploai
df_14_totnghiep['xeploai'].unique()
```

```python
# Chuẩn hóa cột xeploai
df_14_totnghiep.replace(' Khá', 'Khá', inplace=True)
df_14_totnghiep.replace(' Gi?i', 'Giỏi', inplace=True)
df_14_totnghiep.replace([' Trung bình khá', ' Trung bình Khá', ' TB Khá', ' TB khá'],
                        'Trung bình khá', inplace=True)
df_14_totnghiep.replace(' Xu?t s?c', 'Xuất sắc', inplace=True)

# Kiểm tra
df_14_totnghiep['xeploai'].unique()
```

```python
# Cột ngaycapvb
df_14_totnghiep['ngaycapvb'].unique()
```

```python
# Chuẩn hóa cột ngaycapvb
df_14_totnghiep.replace(' 11/10/2019);', ' 11/10/2019', inplace=True)
df_14_totnghiep.replace(' 27-5-2020', ' 27/05/2020', inplace = True)
df_14_totnghiep.replace([' 09/6/2021);', ' 09/6/2021'], ' 09/06/2021', inplace=True)

# Kiểm tra
df_14_totnghiep['ngaycapvb'].unique()

## Làm sao chuyển về datetime
```

```python
df_14_totnghiep.info()
```

```python
df_14_totnghiep.tail(10)
```
