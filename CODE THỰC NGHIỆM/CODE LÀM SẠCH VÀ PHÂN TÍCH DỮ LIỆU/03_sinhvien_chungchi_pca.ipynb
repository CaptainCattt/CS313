## Chuyển đổi kiểu dữ liệu trong các cột

> Văn bản được thụt lề





```python
# clean id column
# change NaN value and strange value ('---', '..') to "0" and then convert column to int64
# remove data with id = '0'

def convert_id(id):
  try:
    id = pd.to_numeric(id)
  except:
    id = np.nan
  return id

df['id'] = df['id'].apply(lambda x: convert_id(x)).fillna(0).astype(np.int64)
df = df[df.id != 0]
```

```python
df['mssv'] = df.mssv.apply(str)
df['mssv'].dtype
```

```python
# clean ngaythi column
# change NaN to NaT and convert column to datetime with format (Year-Month-Year)

df['ngaythi'] = pd.to_datetime(df['ngaythi'], format = '%Y-%m-%d', errors='coerce')
df['ngaythi'].unique()
```

```python
def convert_loaixn(xn):
  try:
    xn = xn.replace("'","")
    xn = xn.replace(" ","")
  except:
    loaixn = np.nan
  return xn

df['loaixn'] = df['loaixn'].apply(lambda x: convert_loaixn(x)) 
df['loaixn'].unique()
```

```python
# CLEAN LOAIXN COLUMN
def convert_loaixn(xn):
  try:
    xn = xn.replace("'","")
    xn = xn.replace(" ","")
  except:
    loaixn = np.nan
  return xn

df['loaixn_2'] = df['loaixn_2'].apply(lambda x: convert_loaixn(x))
df['loaixn_2'].unique()
```

```python
# CLEAN READING COLUMN
# there are 3 type of data: score(float), N3/N4(str), np.nan in column
def convert_reading(reading):
  reading = str(reading)
  reading = reading.replace("'","")
  reading = reading.replace(" ","")
  if reading == 'nan':
    return np.nan
  if 'N' in reading:
    return str(reading)
  else:
    return float(reading)

df['reading'] = df['reading'].apply(lambda x: convert_reading(x))
df['reading'].unique()
```

```python
def convert_writing(writing):
  if writing == ' ' or writing =='':
    return np.nan
  try:
    writing = writing.replace("'","")
    writing = writing.lstrip(' ').rstrip(' ')
    return writing
  except:
    return writing 

df['writing'] = df['writing'].apply(lambda x: convert_writing(x))
df['writing'].unique()
```

```python
# CLEAN DATA OF TONGDIEM COLUMN

def convert_tongdiem(diem):
  try:
    diem = diem.replace("'","")
    diem = diem.lstrip(' ').rstrip(' ')
    return diem
  except:
    return diem

df['tongdiem'] = df.tongdiem.apply(lambda x: convert_tongdiem(x))
df['tongdiem'].unique()
```

```python
# CLEAN DATA OF LYDO COLUMN

def convert_lydo(lydo):
  try:
    lydo = lydo.replace("'",'').lstrip(' ').rstrip(' ')
    return str(lydo)
  except:
    return lydo

df['lydo'] = df['lydo'].apply(lambda x: convert_lydo(x))
df['lydo']
```
