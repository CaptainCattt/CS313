```python
# Chuẩn hóa tên cột (xóa khoảng trắng đầu mỗi tên cột)
df_01_sinhvien = df_01_sinhvien.rename(columns={' masv_tham so dau':"masv_thamsodau",
                                                ' masv': 'masv',
                                                ' namsinh': 'namsinh',
                                                ' gioitinh': 'gioitinh',
                                                ' noisinh': 'noisinh',
                                                ' lopsh': 'lopsh',
                                                ' khoa': 'khoa',
                                                ' hedt': 'hedt',
                                                ' khoahoc': 'khoahoc',
                                                ' chuyennganh2': 'chuyennganh2',
                                                ' tinhtrang': 'tinhtrang',
                                                ' diachi_tinhtp': 'diachi_tinhtp'})
```

```python
# Xóa các dòng không có masv
df_01_sinhvien = df_01_sinhvien.dropna(subset=['mssv'])
```

# 2. Phân tích dữ liệu

```python
df_01_sinhvien = pd.read_csv('/content/drive/MyDrive/MyComputer/Learning/Khai thác dữ liệu và ứng dụng - CS313.N21/edu_data_v2/01_sinhvien.csv')
```

```python
df_01_sinhvien
```

```python
# Thống kê số lượng NA của từng cột
df_01_sinhvien.isna().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
# Phân bố của sinh viên theo khóa học
khoahoc_counts = df_01_sinhvien['khoahoc'].value_counts()
khoahoc_counts
```

```python
plt.bar(khoahoc_counts.index, khoahoc_counts.values)

plt.title('Phân bố sinh viên theo khóa học')
plt.xlabel('Khóa học')
plt.ylabel('Số lượng sinh viên')

plt.show()
```

**Nhận xét:** Số lượng sinh viên tăng dần từ khóa học 8 đến 14. Các khóa học có số lượng sinh viên cao hơn (như khóa học 14 và 13) có thể được cung cấp các chương trình đào tạo hấp dẫn hơn hoặc có nhu cầu cao hơn trong thị trường lao động. Ngược lại, các khóa học có số lượng sinh viên thấp hơn (như khóa học 9 và 8) có thể cần được cải thiện để có thể thu hút thêm sinh viên đăng ký.

> Có thể có nhiều nguyên nhân dẫn đến sự phân bố không đều này, chẳng hạn như chương trình đào tạo không phù hợp với nhu cầu của sinh viên hoặc không đáp ứng được yêu cầu của thị trường lao động.

```python
# Phân bố của sinh viên theo giới tính 
gender_counts = df_01_sinhvien["gioitinh"].value_counts()
gender_counts
```

```python
plt.pie([gender_counts[1], gender_counts[0]], labels=["Nam", "Nữ"], autopct='%1.1f%%')
plt.title("Biểu đồ phân bố sinh viên theo giới tính")
plt.show()
```

**Nhận xét:** Trong tổng số 8295 sinh viên, có 7232 sinh viên là nam (tương đương với 87.2% tổng số sinh viên) và 1063 sinh viên là nữ (tương đương với 12.8% tổng số sinh viên). Tỷ lệ sinh viên nam rất cao so với sinh viên nữ, điều này cho thấy các chương trình đào tạo và các khoa của trường đại học thu hút đa số sinh viên là nam.
> Việc tìm hiểu sự phân bố của sinh viên theo giới tính có thể giúp trường đại học đưa ra những chính sách, chương trình đào tạo phù hợp với từng đối tượng sinh viên.
