# Clean Data

```python
#chuẩn hóa tên cột
df = df.rename(columns = {' listening':'listening',
                         ' reading':'reading',
                         ' total':'total',
                         ' mamh':'mamh',  
                         ' ghichu':'ghichu'})
df.columns
```

```python
df.head(5)
```

```python
df.info()
```

```python
#Xóa các dòng không chứa dữ liệu
df.dropna(axis=0, inplace=True)
```

```python
#chuyển đổi kiểu dữ liệu cột id thành interger

df['id'] = df['id'].apply(int)
df['id'].dtype
```

```python
#chuyển đổi dữ liệu cột listening
def convert_listen(listen):
  if listen == ' NULL':
    return np.nan
  return float(listen)

df['listening'] = df['listening'].apply(lambda x: convert_listen(x))
df['listening'].dtype
```

```python
#chuyển đổi dữ liệu cột reading
def convert_reading(read):
  if read == ' NULL':
    return np.nan
  return float(read)

df['reading'] = df['reading'].apply(lambda x: convert_reading(x))
df['reading'].unique() 
```

```python
#chuyển đổi dữ liệu cột mamh
def convert_mamh(mamh):
  if mamh == ' NULL' or mamh == " ":
    return np.nan
  else:
    mamh = mamh.replace("'",'').lstrip()
    mamh = mamh.replace('EN001', 'ENG01').replace('EN002','ENG02')
    return str(mamh)

df['mamh'] = df['mamh'].apply(lambda x: convert_mamh(x))
df['mamh'].unique()
```

```python
# def filldata_mamh(total, mamh, khoa, hedt):
#   if khoa == 14:
#     if total < 250:
#       return 'AVSC1'
#     elif total <300:
#       mamh == 'AVSC2'
#     elif total < 350:
#       mamh == 'ENG01'
#     elif total < 400:
#       mamh == 'ENG02'
#     elif total < 450:
#       mamh == 'ENG03'
#     elif hedt == 'CQUI':
#       mamh == 'Miễn ENG03'
#     elif total < 500:
#       mamh == 'ENG04'
#     else:
#       mamh == 'ENG05'
# # khoa 13: năm 2018
#   elif khoa == 13:
#     if total < 250:
#       mamh == 'AVSC1'
#     elif total <300:
#       mamh == 'AVSC2'
#     elif total < 350:
#       mamh == 'ENG01'
#     elif total < 400:
#       mamh == 'ENG02'
#     elif total < 450:
#       mamh == 'ENG03'
#     elif total < 500:
#       mamh == 'ENG04'
#     else:
#       mamh == 'ENG05'
# # khoa 12: năm 2017
#   elif khoa == 12:
#     if total >= 96:
#       mamh == 'Anh văn 3'
#     elif total >= 72:
#       mamh == 'Anh văn 2'
#     elif total >=40:
#       mamh == 'Anh văn 1'
#     else:
#       mamh == 'Anh văn bổ túc'
# # khoa 11: năm 2016
#   elif khoa == 11:
#     if total < 26:
#       mamh == 'Anh văn bổ túc'
#     elif total < 55:
#       mamh == 'Anh văn 1'
#     else:
#       mamh == 'Anh văn 2'
#     print(mamh)
#   # return mamh == "100"
  

def filldata_mamh(total, mamh, khoa, hedt):
  if khoa == 14:
    if total < 250:
      return 'AVSC1'
    elif total <300:
      return 'AVSC2'
    elif total < 350:
      return 'ENG01'
    elif total < 400:
      return 'ENG02'
    elif total < 450:
      return 'ENG03'
    elif hedt == 'CQUI':
      return 'Miễn ENG03'
    elif total < 500:
      return 'ENG04'
    else:
      return 'ENG05'
      # khoa 13: năm 2018
  elif khoa == 13:
    if total < 250:
      return 'AVSC1'
    elif total <300:
      return 'AVSC2'
    elif total < 350:
      return 'ENG01'
    elif total < 400:
      return 'ENG02'
    elif total < 450:
      return 'ENG03'
    elif total < 500:
      return 'ENG04'
    else:
      return 'ENG05'
  # khoa 12: năm 2017
  elif khoa == 12:
    if total >= 96:
      return 'Anh văn 3'
    elif total >= 72:
      return 'Anh văn 2'
    elif total >=40:
      return 'Anh văn 1'
    else:
      return 'Anh văn bổ túc'
      # khoa 11: năm 2016
  elif khoa == 11:
    if total < 26:
      return 'Anh văn bổ túc'
    elif total < 55:
      return 'Anh văn 1'
    else:
      return 'Anh văn 2'

df['mamh'] = df[['total','mamh','khoahoc','hedt']].apply(lambda x: filldata_mamh(x['total'], x['mamh'],x['khoahoc'],x['hedt']), axis = 1)
df['mamh'].unique()
```

```python
#chuyển đổi dữ liệu cột ghi chú
def convert_ghichu(ghichu):
  try:
    ghichu = ghichu.replace("'",'').replace(");","").lstrip().rstrip()
    if ghichu == 'NULL':
      return np.nan
    else:
      return ghichu
  except:
    return np.nan

df['ghichu'] = df['ghichu'].apply(lambda x: convert_ghichu(x))
df['ghichu'].unique()
```
