## 1. LÀM SẠCH DỮ LIỆU

```python
import pandas as pd
import numpy as np
```

```python
df_12_baoluu = pd.read_csv("/content/drive/MyDrive/Khai thác dữ liệu và ứng dụng - CS313.N21/Education dataset/12.baoluu.csv")
```

```python
df_12_baoluu.info()
```

```python
# Xóa cột id
df_12_baoluu = df_12_baoluu.iloc[:, 1:]

# Chuẩn hóa tên cột (bỏ khoảng trắng trước tên cột)
df_12_baoluu = df_12_baoluu.rename(columns = {' masv': 'masv',
                                              ' tinhtrang': 'tinhtrang',
                                              ' lydo': 'lydo',
                                              ' hocky': 'hocky',
                                              ' namhoc':'namhoc',
                                              ' soqd': 'soqd',
                                              ' ngayqd': 'ngayqd'})
# Kiểm tra
df_12_baoluu.columns
```

```python
# Xóa 2 dòng không chứa dữ liệu
df_12_baoluu.dropna(axis=0, inplace=True)
```

```python
# Chuyển kiểu dữ liệu của các cột tinhtrang, hocky, namhoc từ float64 sang int
df_12_baoluu['tinhtrang'] = df_12_baoluu['tinhtrang'].astype(int)
df_12_baoluu['hocky'] = df_12_baoluu['hocky'].astype(int)
df_12_baoluu['namhoc'] = df_12_baoluu['namhoc'].astype(int)

# Kiểm tra
df_12_baoluu.info()
```

```python
df_12_baoluu['lydo'].unique()
```

```python
# Chuyển ' ' thành nan
# Bỏ các khoảng trắng đầu chuỗi giá trị

df_12_baoluu.replace(' ', np.nan, inplace=True)
df_12_baoluu.replace(' TN', 'Tốt nghiệp', inplace=True)
df_12_baoluu.replace(' Chứng chỉ Anh văn không đạt --> CNTN tốt nghiệp chuyển sang hệ CQĐT',
                     'Chứng chỉ Anh văn không đạt --> CNTN tốt nghiệp chuyển sang hệ CQĐT',
                     inplace=True)
df_12_baoluu.replace(' QĐ điều chỉnh ngành TN', 'QĐ điều chỉnh ngành TN', inplace=True)
df_12_baoluu.replace(' Tốt nghiệp', 'Tốt nghiệp', inplace=True)

# Kiểm tra
df_12_baoluu['lydo'].unique()
```

```python
df_12_baoluu['soqd'].unique()
## Ví soqd với ngayqd có tính chất tương tự nhau nên không cần sử dụng cột soqd
## Suy ra: Không cần chuẩn hóa cột này
```

```python
df_12_baoluu['ngayqd'].unique()
## Đã OK
```

```python
df_12_baoluu.tail(10)
```

**Nhận xét:** Xóa 2 dòng dữ liệu null và cột index, thu được bảng 12_baoluu gồm 1878 dòng và 7 cột.
